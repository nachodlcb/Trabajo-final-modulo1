## Descripción del Dataset

El dataset incluye las siguientes características:

- **Campaign_ID**: Identificador único de cada campaña.
- **Channel**: Canal de marketing utilizado (e.g., Social Media, Email, TV, News, Radio, Billboard).
- **Product_Category**: Categoría del producto (e.g., Electronics, Fashion, Groceries, Automotive, Health).
- **Region**: Región geográfica en Malasia donde se dirigió la campaña (e.g., Kuala Lumpur, Penang, Malacca, Johor, Sabah, Sarawak).
- **Budget**: Presupuesto asignado en Ringgit malayo.
- **Spend**: Gasto real de la campaña en Ringgit malayo.
- **Impressions**: Número de veces que se vio el anuncio.
- **Clicks**: Número de clics que recibió el anuncio.
- **CTR**: Tasa de clics (Click Through Rate).
- **Conversions**: Número de acciones exitosas (e.g., compras, registros).
- **CR**: Tasa de conversión (Conversion Rate).
- **Revenue**: Ingresos generados por la campaña en Ringgit malayo.
- **ROI**: Retorno de la inversión (Revenue dividido por Spend).
- **Duration**: Duración de la campaña en días.
- **Start_Date**: Fecha de inicio de la campaña.
- **End_Date**: Fecha de finalización de la campaña.
- **Customer_Satisfaction**: Calificación de satisfacción del cliente (1 a 5).
- **Campaign_Effectiveness**: Efectividad de la campaña según el equipo de marketing (1 a 10).
- **Market_Segment**: Segmento de mercado objetivo (Low, Medium, High).

**Nota**: El dataset incluye valores faltantes para simular escenarios del mundo real.

4.2 Manejo de Outliers

- Utiliza gráficos de caja (boxplots) para identificar outliers en variables numéricas.
- Decide cómo manejar los outliers:
  - **Mantenerlos** si son relevantes.
  - **Eliminarlos** si son errores.
  - **Transformarlos** utilizando métodos como logaritmos.

```python
# Comprobamos que las columnas contengan datos numéricos
columns_to_plot = [
    'Budget',
    'Spend',
    'Clicks',
    'Conversions',
    'Revenue',
    'ROI',
]

# Verificar que las columnas existen en el DataFrame y son numéricas
for col in columns_to_plot:
    if col not in df.columns:
        print(f"Columna {col} no encontrada en el DataFrame.")
    elif not pd.api.types.is_numeric_dtype(df[col]):
        print(f"La columna {col} no es numérica.")

# Crear subplots de 2 filas y 3 columnas
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # Crear una cuadrícula de 3x4

# Aplanar el arreglo de ejes para facilitar el acceso
axes = axes.flatten()

# Graficar cada boxplot en su respectivo eje
for ax, col in zip(axes, columns_to_plot):
    ax.boxplot(df[col].dropna())  # Eliminar NaN para evitar errores
    ax.set_title(col)
    ax.set_ylabel('Values')

# Ajustar el layout
plt.tight_layout()
plt.show()
```

```python
# Conclusión sobre los boxplots
conclusions = {}

for col in columns_to_plot:
    # Obtener datos para el análisis
    data = df[col].dropna()
    
    # Calcular estadísticas descriptivas
    median = data.median()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    min_val = data.min()
    max_val = data.max()
    
    # Identificar outliers
    outliers = data[(data < (q1 - 1.5 * iqr)) | (data > (q3 + 1.5 * iqr))]
    
    # Guardar conclusiones
    conclusions[col] = {
        "median": median,
        "q1": q1,
        "q3": q3,
        "min": min_val,
        "max": max_val,
        "outliers": outliers.tolist(),
        "num_outliers": len(outliers)
    }

# Imprimir las conclusiones
for col, stats in conclusions.items():
    print(f"--- Análisis de {col} ---")
    print(f"Mediana: {stats['median']}")
    print(f"Q1: {stats['q1']}, Q3: {stats['q3']} (IQR: {stats['q3'] - stats['q1']})")
    print(f"Valor mínimo: {stats['min']}, Valor máximo: {stats['max']}")
    print(f"Número de valores atípicos: {stats['num_outliers']}")
    print(f"Valores atípicos: {stats['outliers']}\n")
```

```python
numerical_vars = ['Budget', 'Spend', 'Impressions', 'Clicks', 'Conversions', 'CTR', 'ROI']

# Función para eliminar outliers utilizando IQR
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Aplicar la función para eliminar outliers
df = remove_outliers_iqr(df, numerical_vars)

# Mostrar el DataFrame limpio
df
```

```python
# Lista de columnas a analizar
numeric_cols = ['Budget', 'Spend', 'Impressions', 'Clicks', 'CTR', 
                'Conversions', 'CR', 'Revenue', 'ROI', 'Duration', 
                'Customer_Satisfaction', 'Campaign_Effectiveness']

# Configuración de los gráficos
plt.figure(figsize=(15, 12))

for i, col in enumerate(numeric_cols):
    plt.subplot(4, 3, i + 1)  # Ajusto el número de filas y columnas según sea necesario
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histograma y KDE de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
# Descripción de los histogramas y KDE
def describe_histograms(df, columns):
    descriptions = {}
    for col in columns:
        descriptions[col] = {
            'mean': df[col].mean(),
            'median': df[col].median(),
            'std': df[col].std(),
            'min': df[col].min(),
            'max': df[col].max(),
            'skewness': df[col].skew(),
            'kurtosis': df[col].kurt()
        }
    return descriptions

# Llamar a la función y mostrar las descripciones
histogram_descriptions = describe_histograms(df, numeric_cols)
for col, stats in histogram_descriptions.items():
    print(f"--- Descripción de {col} ---")
    for stat, value in stats.items():
        print(f"{stat}: {value}")
    print("\n")

```

```python
# Lista de variables categóricas
categorical_vars = ['Channel', 'Product_Category', 'Region', 'Market_Segment']

# Crear un diccionario para almacenar las tablas de contingencia
contingency_tables = {}

# Crear combinaciones de variables categóricas
for i, var1 in enumerate(categorical_vars):
    for j, var2 in enumerate(categorical_vars):
        if var1 != var2:  # Asegurarse de que no sea la misma variable
            contingency_table = pd.crosstab(df[var1], df[var2], margins=True)
            contingency_tables[f'{var1} vs {var2}'] = contingency_table
            
            # Imprimir la tabla de contingencia
            print(f"Tabla de Contingencia: {var1} vs {var2}")
            print(contingency_table)
            print("\n")
            
            # Calcular porcentajes
            percentage_table = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100
            print(f"Porcentajes: {var1} vs {var2}")
            print(percentage_table)
            print("\n")
```

```python
# Lista de variables numéricas específicas
numerical_vars = ['Budget', 'Spend', 'Impressions', 'Clicks', 'Conversions', 'CTR', 'ROI']

# Rellenar o eliminar valores nulos en las variables numéricas seleccionadas
df[numerical_vars] = df[numerical_vars].fillna(0)  # O elige otro método de imputación

# Calcular la matriz de correlación
correlation_matrix = df[numerical_vars].corr()

# Configurar la figura
plt.figure(figsize=(10, 8))

# Crear un heatmap de la matriz de correlación
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar_kws={'label': 'Correlación'})

# Títulos y etiquetas
plt.title('Matriz de Correlación de Variables Numéricas Específicas')
plt.xlabel('Variables')
plt.ylabel('Variables')

plt.tight_layout()
plt.show()

# Descripción de la matriz de correlación
for var in numerical_vars:
    correlations = correlation_matrix[var].drop(var)  # Excluye la correlación consigo mismo
    high_corr = correlations[correlations.abs() > 0.5]  # Filtra correlaciones altas
    low_corr = correlations[correlations.abs() < 0.1]  # Filtra correlaciones bajas
    
    print(f"\nAnálisis de la variable: {var}")
    print(f"Correlaciones altas (|r| > 0.5):\n{high_corr}\n")
    print(f"Correlaciones bajas (|r| < 0.1):\n{low_corr}\n")
```

```python
numerical_vars = ['Budget', 'Spend', 'Impressions', 'Revenue', 'Clicks', 'Conversions', 'CTR', 'ROI']

# Filtrar el DataFrame
data = df[numerical_vars]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

inertia = []
K_range = range(1, 20)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.xticks(K_range)
plt.grid()
plt.show()
```

```python
# Decido usar 7 clusters
optimal_k = 7
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

plt.figure(figsize=(10, 6))
sns.scatterplot(df, x='Budget', y='Spend', hue='Cluster', palette='deep', s=100)
plt.title('Clusters de Campañas (K-means con 7 Clusters)')
plt.xlabel('Budget')
plt.ylabel('Spend')
plt.legend(title='Cluster')
plt.show()

cluster_summary = df.groupby('Cluster')[numerical_vars].mean()
print(cluster_summary)
```

```python
# Generar el gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Budget', y='ROI', hue='Market_Segment', palette='viridis', alpha=0.7)
plt.title('Relación entre Budget y ROI por Segmento de Mercado')
plt.xlabel('Budget')
plt.ylabel('ROI')
plt.legend(title='Segmento de Mercado')
plt.grid()
plt.show()

# Descripción automática del gráfico
def describe_scatter_plot(df):
    description = {
        'Budget': {
            'mean': df['Budget'].mean(),
            'median': df['Budget'].median(),
            'min': df['Budget'].min(),
            'max': df['Budget'].max(),
            'std': df['Budget'].std()
        },
        'ROI': {
            'mean': df['ROI'].mean(),
            'median': df['ROI'].median(),
            'min': df['ROI'].min(),
            'max': df['ROI'].max(),
            'std': df['ROI'].std()
        },
        'Market_Segment': df['Market_Segment'].value_counts().to_dict()
    }
    
    print("Descripción del gráfico de dispersión:")
    print("Estadísticas de Budget:")
    for key, value in description['Budget'].items():
        print(f"{key}: {value}")
    
    print("\nEstadísticas de ROI:")
    for key, value in description['ROI'].items():
        print(f"{key}: {value}")

    print("\nDistribución de Segmentos de Mercado:")
    for segment, count in description['Market_Segment'].items():
        print(f"{segment}: {count} registros")

# Llamar a la función
describe_scatter_plot(df)
```
